Позволяет быстро определить разброс значений для каждого параметра автомобиля в наборе данных. Вывод результатов поможет вам лучше понять диапазоны значений каждого параметра и провести более детальный анализ данных:

*   Для каждого параметра (displacement, mpg, cylinders, horsepower, weight, acceleration, model_year, origin) вычисляются максимальные и минимальные значения с помощью функций np.max() и np.min() из библиотеки NumPy.

*   Затем выводятся результаты вычислений с помощью print() для каждого параметра, чтобы отобразить максимальное и минимальное значение в каждом столбце данных.



```python
max_displacement = np.max(displacement)
min_displacement = np.min(displacement)

print("Максимальное значение в столбце displacement:", max_displacement)
print("Минимальное значение в столбце displacement:", min_displacement)

max_mpg = np.max(mpg)
min_mpg = np.min(mpg)

print("Максимальное значение в столбце mpg:", max_mpg)
print("Минимальное значение в столбце mpg:", min_mpg)

max_cylinders = np.max(cylinders)
min_cylinders = np.min(cylinders)

print("Максимальное значение в столбце cylinders:", max_cylinders)
print("Минимальное значение в столбце cylinders:", min_cylinders)

max_horsepower = np.max(horsepower)
min_horsepower = np.min(horsepower)

print("Максимальное значение в столбце horsepower:", max_horsepower)
print("Минимальное значение в столбце horsepower:", min_horsepower)

max_weight = np.max(weight)
min_weight = np.min(weight)

print("Максимальное значение в столбце weight:", max_weight)
print("Минимальное значение в столбце weight:", min_weight)

max_acceleration = np.max(acceleration)
min_acceleration = np.min(acceleration)

print("Максимальное значение в столбце acceleration:", max_acceleration)
print("Минимальное значение в столбце acceleration:", min_acceleration)

max_model_year = np.max(model_year)
min_model_year = np.min(model_year)

print("Максимальное значение в столбце model_year:", max_model_year)
print("Минимальное значение в столбце model_year:", min_model_year)

max_origin = np.max(origin)
min_origin = np.min(origin)

print("Максимальное значение в столбце origin:", max_origin)
print("Минимальное значение в столбце origin:", min_origin)
```

Загружаем данные из файла "auto-mpg.data-original", используя NumPy, и затем выводит их на экран:

*   Импортируется библиотека NumPy под псевдонимом np;
*   Функция np.genfromtxt() используется для загрузки данных из файла "auto-mpg.data-original": delimiter="," указывает, что данные в файле разделены запятыми и dtype=None означает, что тип данных будет определен автоматически;
*   Загруженные данные сохраняются в переменной data;
*   Данные data выводятся на экран с помощью функции print().



```python
import numpy as np

data = np.genfromtxt("auto-mpg.data-original", delimiter=",", dtype=None)

print(data)
```

Визуализировает зависимость переменной MPG от различных параметров автомобилей и провести исследование данных с помощью графиков:

*   Данные загружаются из массива data в отдельные списки для каждого параметра автомобиля: displacement, mpg, cylinders, horsepower, weight, acceleration, model_year, origin.

*   Далее создаются несколько фигур (графиков) с помощью plt.figure() для каждого параметра: displacement, cylinders, horsepower, weight, acceleration, model_year, origin.

*   Для каждого параметра строятся точечные графики, где по оси X отложены значения параметра, а по оси Y - значения MPG. Например, для параметра displacement и mpg создаются точечные графики с использованием plt.plot() и различных маркеров и цветов.

*   Для каждого графика добавляются подписи осей X и Y с помощью plt.xlabel() и plt.ylabel(), а также легенда с помощью plt.legend() для обозначения разных диапазонов данных.

*   Наконец, с помощью plt.tight_layout() обеспечивается компактное расположение графиков на холсте, и plt.show() отображает все созданные графики



```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#Данные из отдельных столбцов
displacement = []
mpg = []
cylinders = []
horsepower = []
weight = []
acceleration = []
model_year = []
origin = []

for dot in data:
  displacement.append(dot[0])
  mpg.append(dot[1])
  cylinders.append(dot[2])
  horsepower.append(dot[3])
  weight.append(dot[4])
  acceleration.append(dot[5])
  model_year.append(dot[6])
  origin.append(dot[7])

plt.figure(1)
mpg_displacement = plt.plot(displacement[:100], mpg[:100], 'ro', label='mpg_displacement')
mpg_displacement = plt.plot(displacement[100:200], mpg[100:200], 'g^', label='mpg_displacement')
mpg_displacement = plt.plot(displacement[200:300], mpg[200:300], 'bs', label='mpg_displacement')
mpg_displacement = plt.plot(displacement[300:398], mpg[300:398], 'go', label='mpg_displacement')

plt.xlabel("Displacement")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(2)
mpg_cylinders = plt.plot(cylinders[:100], mpg[:100], 'ro', label='mpg_cylinders')
mpg_cylinders = plt.plot(cylinders[100:200], mpg[100:200], 'g^', label='mpg_cylinders')
mpg_cylinders = plt.plot(cylinders[200:300], mpg[200:300], 'bs', label='mpg_cylinders')
mpg_cylinders = plt.plot(cylinders[300:398], mpg[300:398], 'go', label='mpg_cylinders')

plt.xlabel("Cylinders")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(3)
mpg_horsepower = plt.plot(horsepower[:100], mpg[:100], 'ro', label='mpg_horsepower')
mpg_horsepower = plt.plot(horsepower[100:200], mpg[100:200], 'g^', label='mpg_horsepower')
mpg_horsepower = plt.plot(horsepower[200:300], mpg[200:300], 'bs', label='mpg_horsepower')
mpg_horsepower = plt.plot(horsepower[300:398], mpg[300:398], 'go', label='mpg_horsepower')

plt.xlabel("Horsepower")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(4)
mpg_weight = plt.plot(weight[:100], mpg[:100], 'ro', label='mpg_weight')
mpg_weight = plt.plot(weight[100:200], mpg[100:200], 'g^', label='mpg_weight')
mpg_weight = plt.plot(weight[200:300], mpg[200:300], 'bs', label='mpg_weight')
mpg_weight = plt.plot(weight[300:398], mpg[300:398], 'go', label='mpg_weight')

plt.xlabel("Weight")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(5)
mpg_acceleration = plt.plot(acceleration[:100], mpg[:100], 'ro', label='mpg_acceleration')
mpg_acceleration = plt.plot(acceleration[100:200], mpg[100:200], 'g^', label='mpg_acceleration')
mpg_acceleration = plt.plot(acceleration[200:300], mpg[200:300], 'bs', label='mpg_acceleration')
mpg_acceleration = plt.plot(acceleration[300:398], mpg[300:398], 'go', label='mpg_acceleration')

plt.xlabel("Acceleration")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(6)
mpg_model_year = plt.plot(model_year[:100], mpg[:100], 'ro', label='mpg_model_year')
mpg_model_year = plt.plot(model_year[100:200], mpg[100:200], 'g^', label='mpg_model_year')
mpg_model_year = plt.plot(model_year[200:300], mpg[200:300], 'bs', label='mpg_model_year')
mpg_model_year = plt.plot(model_year[300:398], mpg[300:398], 'go', label='mpg_model_year')

plt.xlabel("Model year")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.figure(7)
mpg_origin = plt.plot(origin[:100], mpg[:100], 'ro', label='mpg_origin')
mpg_origin = plt.plot(origin[100:200], mpg[100:200], 'g^', label='mpg_origin')
mpg_origin = plt.plot(origin[200:300], mpg[200:300], 'bs', label='mpg_origin')
mpg_origin = plt.plot(origin[300:398], mpg[300:398], 'go', label='mpg_origin')

plt.xlabel("Origin")
plt.ylabel('Mpg')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
