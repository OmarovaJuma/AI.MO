**Построение классификатора с заданием K (количество ближайших соседей) пользователем**

Разделение данных на обучающую и тестовую выборки:

> Данные d разделяются на обучающую выборку X_train и целевую переменную y_train.

> Обучающая выборка X_train содержит числовые характеристики (независимые переменные), такие как показатель преломления (RI), содержание натрия (Na) и т. д.

Целевая переменная y_train содержит классы стекла (зависимая переменная).
Запрос количества ближайших соседей у пользователя:

> Пользователь вводит количество ближайших соседей K, которое будет использоваться в алгоритме K ближайших соседей (KNN).

Создание классификатора KNN:

> Создается классификатор KNN с указанным количеством ближайших соседей K.

Обучение классификатора:

> Классификатор KNN обучается на обучающей выборке X_train и y_train.

Подготовка тестовых данных:

> Подготавливаются три тестовых образца X_test_1, X_test_2 и X_test_3, каждый из которых представляет собой набор числовых характеристик.

Предсказание классов тестовых данных:

> Обученный классификатор KNN используется для предсказания классов тестовых данных.

> Переменные target_1, target_2 и target_3 содержат предсказанные классы для соответствующих тестовых образцов.

Вывод результата:

> Предсказанные классы для тестовых данных выводятся на консоль.

В результате код использует алгоритм KNN для классификации типов стекла на основе их числовых характеристик. Пользователь может указать количество ближайших соседей, которое будет использоваться в алгоритме, и затем код предсказывает классы для новых тестовых образцов.

```python
# Разделение данных на обучающую и тестовую выборки
X_train = d[['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']]
y_train = d['Type_of_glass']

# Запрос количества ближайших соседей у пользователя
K = int(input("Введите количество ближайших соседей: "))

# Создание классификатора
knn = KNeighborsClassifier(n_neighbors=K)

# Обучение классификатора
knn.fit(X_train, y_train)

# Подготовка тестовых данных
X_test_1 = np.array([[1.51743,13.30,3.60,1.14,73.09,0.58,8.17,0.00,0.00]])
X_test_2 = np.array([[1.52101,13.64,4.49,1.10,71.85,0.00,1.51,0.58,0.00]])
X_test_3 = np.array([[1.51363,14.29,3.60,0.95,72.61,0.00,0.77,0.61,0.00]])

# Предсказание классов тестовых данных
target_1 = knn.predict(X_test_1)
target_2 = knn.predict(X_test_2)
target_3 = knn.predict(X_test_3)

# Вывод результата
print("Предсказанный класс для тестовых данных 1:", target_1)
print("Предсказанный класс для тестовых данных 2:", target_2)
print("Предсказанный класс для тестовых данных 3:", target_3)
```

**Вычислить оценки hold-out для различных значений K, а также для различных долей обучающей и тестирующей подвыборок.**

Разделение данных на обучающую и тестовую выборки:

> Данные d разделяются на обучающую выборку X_train и y_train и тестовую выборку X_holdout и y_holdout.

> Разделение выполняется с использованием функции train_test_split из библиотеки sklearn.model_selection с указанием доли тестовой выборки test_size и фиксированного случайного состояния random_state.

Перебор параметров K и доли обучающей выборки:

> Код перебирает различные значения количества ближайших соседей K (от 1 до 10) и доли обучающей выборки (10%, 20% и 30%).

Обучение и оценка классификатора KNN:

> Для каждой комбинации K и доли обучающей выборки создается классификатор KNN и обучается на обучающей выборке.

> Затем классификатор используется для предсказания классов тестовой выборки, и вычисляется точность (доля правильно предсказанных классов).

Сохранение результатов:

> Точность сохраняется в списке accuracies.

Вывод результатов:

> Результаты выводятся в виде таблицы, где строки соответствуют доле обучающей выборки, а столбцы соответствуют значению K.

> Для каждой комбинации параметров выводится точность.

В результате код выполняет перекрестную проверку для алгоритма KNN, чтобы определить оптимальное количество ближайших соседей и долю обучающей выборки для набора данных.

```python
# Разделение данных на обучающую и тестовую выборки
test_sizes = [0.1, 0.2, 0.3]
accuracies = []

for test_size in test_sizes:
    for k in range(1, 11):
        X_train, X_holdout, y_train, y_holdout = train_test_split(d[['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']],
                                                                  d['Type_of_glass'],
                                                                  test_size=test_size,
                                                                  random_state=17)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        knn_pred = knn.predict(X_holdout)
        accuracies.append(accuracy_score(y_holdout, knn_pred))

# Вывод результатов
print('Оценки hold-out для различных значений K и долей обучающей и тестирующей подвыборок:')
for i, test_size in enumerate(test_sizes):
    print(f'Доля обучающей выборки: {1 - test_size}')
    for j in range(1, 11):
        print(f'K = {j}: {accuracies[i * 10 + j - 1]}')
    print()
```

**Вычислить оценки cross validation для различных значений K, а также для различных значений fold (количество подмножеств при кросс-валидации).**

Создание списка значений K:

> Создается список k_list со значениями K от 1 до 49.
Инициализация списка для хранения оценок кросс-валидации:

> Создается пустой список cv_scores для хранения средних оценок кросс-валидации для каждого значения K.

Перебор значений K:

> Код перебирает значения K в списке k_list.

Обучение и оценка классификатора KNN:

> Для каждого значения K создается классификатор KNN и обучается на данных d.iloc[:, 0:4] (первые четыре столбца данных) с использованием целевого признака d['Type_of_glass'].

> Оценки кросс-валидации вычисляются с использованием 10-кратной кросс-валидации и метрики точности.

> Среднее значение оценок кросс-валидации добавляется в список cv_scores.

Вычисление ошибки средней квадратичной ошибки (MSE):

> Список MSE создается путем вычитания средних оценок кросс-валидации из 1. MSE представляет собой ошибку классификации.

Определение оптимального значения K:

> Индекс минимального значения в списке MSE находится, и соответствующее ему значение K в списке k_list является оптимальным значением K.

> Список all_k_min создается для хранения всех значений K, соответствующих минимальной MSE.

Вывод результатов:

> Оптимальное значение K выводится на консоль.

В результате код выполняет перекрестную проверку для алгоритма KNN с различными значениями K, чтобы определить оптимальное количество ближайших соседей для набора данных. Он также вычисляет MSE для каждого значения K и строит график зависимости MSE от K.

```python
#Создание списка значений K
k_list = list(range(1, 50))

#Инициализация списка для хранения оценок кросс-валидации
cv_scores = []

#Цикл по значениям K
for k in k_list:
    # Создание экземпляра классификатора K ближайших соседей с заданным количеством соседей
    knn = KNeighborsClassifier(n_neighbors=k)
    # Вычисление оценок кросс-валидации с использованием 10-кратной кросс-валидации и метрики точности
    scores = cross_val_score(knn, d.iloc[:, 0:4], d['Type_of_glass'], cv=10, scoring='accuracy')
    # Добавление среднего значения оценок в список
    cv_scores.append(scores.mean())

#Вычисление ошибки средней квадратичной ошибки (MSE)
MSE = [1 - x for x in cv_scores]
optimal_k = k_list[MSE.index(min(MSE))]

#Построение графика зависимости MSE от K
plt.plot(k_list, MSE)
plt.xlabel('Количество соседей (K)')
plt.ylabel('Ошибка классификации (MSE)')
plt.show()

#Определение оптимального значения K
k_min = min(MSE)

#Создание списка всех значений K, соответствующих минимальной MSE
all_k_min = []
for i in range(len(MSE)):
    if MSE[i] <= k_min:
        all_k_min.append(k_list[i])

#Вывод оптимального значения K
print('Оптимальное значение K:', all_k_min)
```

**Вычислить оптимальное значение K. Обосновать выбор. Продемонстрировать использование полученного классификатора.**

Оптимальное значение K зависит от конкретного набора данных и задачи классификации. Однако в целом более высокие значения K приводят к более гладкой границе решения, в то время как более низкие значения K приводят к более сложной границе решения.

Для данного набора данных мы вычисляем оценки кросс-валидации для различных значений K и строим график зависимости MSE от K. Оптимальное значение K определяется как значение, соответствующее минимальной MSE.

Разделение данных на обучающий и тестовый наборы:

Этот код разделяет данные на обучающий и тестовый наборы. Функция train_test_split случайным образом делит данные на две части: обучающий набор (X_train и y_train) и тестовый набор (X_test и y_test). В этом случае размер тестового набора составляет 20% от общего набора данных.

Создание классификатора KNN:

Этот код создает экземпляр классификатора KNN с оптимальным значением параметра n_neighbors, которое было определено ранее. Классификатор KNN предсказывает метку класса для новой точки данных, основываясь на метках класса K наиболее похожих точек данных в обучающем наборе.

Обучение классификатора:

Этот код обучает классификатор KNN на обучающем наборе. Во время обучения классификатор изучает отношения между признаками и метками классов в обучающем наборе.

Предсказание меток классов:

Этот код использует обученный классификатор KNN для предсказания меток классов для точек данных в тестовом наборе.

Оценка точности:

Этот код оценивает точность классификатора на тестовом наборе, сравнивая предсказанные метки классов (y_pred) с истинными метками классов (y_test). Функция accuracy_score возвращает значение точности, которое представляет собой долю правильно предсказанных меток классов.

```python
# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(d.iloc[:, 0:4], d['Type_of_glass'], test_size=0.2)

# Создание экземпляра классификатора KNN с оптимальным значением K
knn = KNeighborsClassifier(n_neighbors=optimal_k)

# Обучение классификатора на обучающем наборе
knn.fit(X_train, y_train)

# Предсказание меток классов для тестового набора
y_pred = knn.predict(X_test)

# Оценка точности классификатора на тестовом наборе
accuracy = accuracy_score(y_test, y_pred)
print('Точность классификатора на тестовом наборе:', accuracy)
```



### 1. Особенности основных методов метрической классификации: метод ближайшего соседа и метод k ближайших соседей.

- **Метод ближайшего соседа (1-NN)**: Этот метод классифицирует объект, основываясь на классе его ближайшего соседа в обучающей выборке. Он прост в реализации, но чувствителен к шуму и выбросам, так как может принять решение на основе одного примера, который может быть аномалией.

- **Метод k ближайших соседей (k-NN)**: Этот метод классифицирует объект на основе классов k ближайших соседей. Значение k может варьироваться, и оно влияет на устойчивость модели к шуму. При этом учитывается большее количество соседей, что делает классификацию более надежной и устойчивой.

### 2. Основные принципы и этапы реализации метода kNN.

1. **Выбор значения k**: Определение количества соседей, которые будут учитываться при классификации.
2. **Вычисление расстояний**: Для каждого объекта в обучающей выборке вычисляются расстояния до нового объекта. Обычно используется евклидово расстояние, но могут применяться и другие метрики.
3. **Сортировка соседей**: Соседи сортируются по возрастанию расстояния.
4. **Голосование**: Определяется класс нового объекта на основе классов k ближайших соседей. Обычно используется простое голосование, но могут применяться и взвешенные методы.
5. **Классификация**: Новый объект классифицируется в класс, который получил наибольшее количество голосов.

### 3. Принцип выбора количества соседних объектов.

Выбор значения k является критически важным. Малое значение k может привести к переобучению (чувствительность к шуму), а слишком большое значение может привести к недообучению (смешивание классов). Обычно выбирают k через кросс-валидацию, тестируя различные значения и выбирая то, которое дает наилучшие результаты на валидационном наборе данных.

### 4. Метод парзеновского окна.

Метод парзеновского окна (или парзеновская оценка плотности) — это метод, который использует окна (или ядра) для оценки плотности распределения данных. Вместо того чтобы классифицировать объект на основе фиксированного числа соседей, он оценивает плотность данных в окрестности объекта. Объекты, находящиеся ближе к целевому объекту, имеют больший вес в оценке плотности.

### 5. Принцип метода потенциальных функций.

Метод потенциальных функций основывается на использовании функций, которые моделируют влияние соседних объектов на целевой объект. Каждому объекту присваивается "потенциал", который зависит от его расстояния до целевого объекта. Классификация осуществляется на основе суммирования потенциалов всех объектов, и класс с наибольшим потенциалом становится предсказанным классом для целевого объекта.

### 6. Параметры, оптимизируемые в методах kNN.

В методах kNN оптимизируются следующие параметры:

- **Количество соседей (k)**: Оптимальное значение k, которое минимизирует ошибку классификации.
- **Метрика расстояния**: Выбор подходящей метрики (евклидова, манхэттенская и т.д.) для вычисления расстояний между объектами.
- **Вес соседей**: В некоторых случаях используется взвешенное голосование, где ближние соседи имеют больший вес, чем дальние.
